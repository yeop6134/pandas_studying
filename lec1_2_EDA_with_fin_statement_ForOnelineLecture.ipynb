## 연산 관련 pandas built-in 함수

- axis란?
- 연산은 기본적으로 "axis를 변형(줄이거나 늘리는)하는 방식" 으로 진행된다.
- "axis를 변형하는 방식으로 작동한다"의 원리는 aggregation 연산자 (여러개의 값을 하나의 값으로 변형하는 연산 ex: sum())에 대해서만 적용이 가능하다. (예를 들어 빼기 (subtract)에서는 적용되지 않는다.

```python
#### numpy로 맛보기

import numpy as np

a = np.array([1,2,3])
b = np.array([1,2,3])
a
b
```

```python
a + b
```

```python
np.sum(
    [a,b],
    axis=0  # 2차원 데이터에서는 세로로 더해진다고 생각해도 됨
)           # axis=0 (0,1)의 구분을 줄이는 방법!
```

```python
np.sum(
    [a,b],
    axis=1  # 2차원 데이터에서는 가로로 더해진다고 생각해도 됨
)           # axis=1(0,1,2)의 구분을 줄이는 방법! 
```

```python
[      0  1  2  axis=1
   0  [1, 2, 2],
   1  [1, 2, 2]
 axis=0
]
```

```python
 axis=0 --->       0            1           # 가장 바깥쪽 index가 axis=0
 axis=1 --->    0  1  2      0  1  2
            [  [1, 2, 2],   [1, 2, 2] ]
```

```python
 axis=0 --->              0                        1                 # 가장 바깥쪽 index가 axis=0
 axis=1 --->      0       1     2          0       1      2
 axis=2 --->     0  1    0, 1    0        0, 1    0, 1    0, 1
            [  [[1, 2], [2, 4], [2]],   [[1, 2], [2, 2], [2, 2]] ]
```
